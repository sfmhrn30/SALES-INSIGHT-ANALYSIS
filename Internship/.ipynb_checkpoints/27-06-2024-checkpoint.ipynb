```python
import pandas as pd 

# Load data 
manpower =pd.read_csv('datasets copies/Man_Power.csv') 
sales_plan = pd.read_csv('datasets copies/Sales_Plan.csv') 
outsourced_smh= pd.read_csv('datasets copies/Outsourced_SMH.csv') 
inhouse_smh= pd.read_csv('datasets copies/Inhouse_SMH.csv')
dispatches = pd.read_csv('datasets copies/Dispatches_finyears.csv')
```

```python
print(manpower.columns)
print(sales_plan.columns)
print(outsourced_smh.columns)
print(inhouse_smh.columns)
print(dispatches.columns)
```

```python
dataDic={}
dataDic['financial_year']=manpower['financial_year']
dataDic['man_power']=manpower['Man_Power']
dataDic['sales']=sales_plan['Sales']
dataDic['outsourced_smh']=outsourced_smh['smh_lacs']

annual_smh_dispatches =dispatches.groupby('financial_year')['INC_SMH'].sum().reset_index()
dataDic['annual_smh_dispatches']=annual_smh_dispatches['INC_SMH']

#calculating efficiency
years=['2021_22', '2022_23', '2023_24', '2024_25']
efficiency=[]
annual_inhouse_smh=[]
for i in years:
    smhYear=inhouse_smh['SMH_'+i].sum()
    bookYear=inhouse_smh['BOOK_'+i].sum()
    efficiency.append((smhYear/bookYear)*100)
    annual_inhouse_smh.append(smhYear)
dataDic['efficiency']=efficiency
dataDic['annual_inhouse_smh']=annual_inhouse_smh
print(dataDic)
```

```python
data=pd.DataFrame(dataDic)
print(data)
```

```python
data.to_csv('merge1.csv', index=False)
```
